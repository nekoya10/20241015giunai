```python
import matplotlib.pyplot as plt

labels = ['Nokia','Samaung','Apple','Lumia']
values = [20, 30, 45, 10]
colors = ['yellow','green','red','blue']
explode = (0.3, 0, 0, 0)#設定偏移的距離 順序分別是(Nokia,Samsun,Apple,Lumia)

figure = plt.subplots()#製作figure 
axes = plt.subplot(1,1,1)#在figure上製作一個axes

#autopct:顯示百分比 shadow:顯示影子 startangle轉換角度
axes.pie(values,
        explode=explode,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        shadow=True,
        startangle=180)
plt.show()
```
